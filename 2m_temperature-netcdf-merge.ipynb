### Convert netCDF to CSV

```python
# Convert cftime → python datetime
times_raw = nc.num2date(temperature_2020['valid_time'][:], temperature_2020['valid_time'].units)
times = [dt.datetime(t.year, t.month, t.day, t.hour, t.minute, t.second)
         for t in times_raw]

lat = temperature_2020['latitude'][:]
lon = temperature_2020['longitude'][:]
t2m = temperature_2020['t2m'][:]

# Choose your exact location
target_lat = 52.23
target_lon = 21.01

# Find nearest grid cell
ilat = np.argmin(np.abs(lat - target_lat))
ilon = np.argmin(np.abs(lon - target_lon))

# Extract hourly values
t2m_point = t2m[:, ilat, ilon]


temperature_2020_df = pd.DataFrame({
    "time": times,
    "lat": lat[ilat],
    "lon": lon[ilon],
    "t2m": t2m_point
})

display(temperature_2020_df)


# save df
temperature_2020_df.to_parquet("../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2020.parquet")
```

```python
## now i have clouds 2020 dataset with one lat x long 
## here i will save another dataset with all locations and times (for better visualization)

times_raw = nc.num2date(temperature_2020['valid_time'][:], temperature_2020['valid_time'].units)

times = [
    dt.datetime(t.year, t.month, t.day, t.hour, t.minute, t.second)
    for t in times_raw
]


lat = temperature_2020['latitude'][:]      # shape (25,)
lon = temperature_2020['longitude'][:]     # shape (43,)
t2m = temperature_2020['t2m'][:]           # shape (8784, 25, 43)


# Build a tidy DataFrame for all grid points

rows = []

for ti, t in enumerate(times):
    for i, la in enumerate(lat):
        for j, lo in enumerate(lon):
            rows.append([t, float(la), float(lo), float(t2m[ti, i, j])])

temperature_2020_lat_long_full_df = pd.DataFrame(rows, columns=["time", "lat", "lon", "t2m"])


display(temperature_2020_lat_long_full_df)


# save df_all
temperature_2020_lat_long_full_df.to_parquet("../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_all_2020.parquet")
```

## Merging 2020 - 2025 parquet data

```python
file_path_2020 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2020.parquet')
file_path_2021 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2021.parquet')
file_path_2022 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2022.parquet')
file_path_2023 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2023.parquet')
file_path_2024 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2024.parquet')
file_path_2025 = Path('../Data/raw-data/nc-climate-dataset/2m_temperature/processed/2m_temperature_2025.parquet')
```

```python
temperature_2020 = pd.read_parquet(file_path_2020)
temperature_2021 = pd.read_parquet(file_path_2021)
temperature_2022 = pd.read_parquet(file_path_2022)
temperature_2023 = pd.read_parquet(file_path_2023)
temperature_2024 = pd.read_parquet(file_path_2024)
temperature_2025 = pd.read_parquet(file_path_2025)
```

```python
# consolidate those files from 2020 to 2025

temperature_2020_2025 = pd.concat(
    [
        temperature_2020,
        temperature_2021,
        temperature_2022,
        temperature_2023,
        temperature_2024,
        temperature_2025
    ], 
    ignore_index=True
)

display(temperature_2020_2025)
```

```python
import matplotlib.dates as mdates

plt.rcParams['font.family'] = 'Times New Roman'

plt.figure(figsize=(15, 8))
plt.plot(
    temperature_2020_2025['t2m'],
    label='2 metre temperature',
    linewidth=1.2,
    color='navy'
)

plt.title('2 metre temperature from 2020 to 2025 (Hourly)', fontsize=22)
plt.xlabel('Date and Time', fontsize=18)
plt.ylabel('2 metre temperature', fontsize=18)

# Monthly ticks
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval = 3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))

plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

plt.legend(fontsize=18)

plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.5, color='black')

plt.tight_layout()

plt.savefig(
    "2m_temperature_2020_2025.png",
    dpi=300,               # high resolution
    bbox_inches='tight',   # trims extra whitespace
    facecolor='white'      # ensures white background
)

plt.show()
```
