### Convert netCDF to CSV

```python
# Convert cftime → python datetime
times_raw = nc.num2date(wind_2020['valid_time'][:], wind_2020['valid_time'].units)
times = [dt.datetime(t.year, t.month, t.day, t.hour, t.minute, t.second)
         for t in times_raw]

# Read lat/lon
lat = wind_2020['latitude'][:]
lon = wind_2020['longitude'][:]

# Read wind components
u100 = wind_2020['u100'][:]   # shape: (time, lat, lon)
v100 = wind_2020['v100'][:]

# Choose your exact location
target_lat = 52.23
target_lon = 21.01

# Find nearest grid cell
ilat = np.argmin(np.abs(lat - target_lat))
ilon = np.argmin(np.abs(lon - target_lon))

# Extract hourly values
u_point = u100[:, ilat, ilon]
v_point = v100[:, ilat, ilon]

# Build dataframe
wind_2020_df = pd.DataFrame({
    "time": times,
    "lat": lat[ilat],
    "lon": lon[ilon],
    "u100": u_point,
    "v100": v_point
})

display(wind_2020_df)

# Save
wind_2020_df.to_parquet("../Data/raw-data/nc-climate-dataset/wind/processed/wind_2020.parquet")
```

```python
# Convert cftime → python datetime
times_raw = nc.num2date(wind_2020['valid_time'][:], wind_2020['valid_time'].units)

times = [
    dt.datetime(t.year, t.month, t.day, t.hour, t.minute, t.second)
    for t in times_raw
]

# Read grid
lat = wind_2020['latitude'][:]      # shape (25,)
lon = wind_2020['longitude'][:]     # shape (43,)

# Read wind components
u100 = wind_2020['u100'][:]         # shape (8784, 25, 43)
v100 = wind_2020['v100'][:]         # shape (8784, 25, 43)

# Build tidy DataFrame for all grid points
rows = []

for ti, t in enumerate(times):
    for i, la in enumerate(lat):
        for j, lo in enumerate(lon):
            rows.append([
                t,
                float(la),
                float(lo),
                float(u100[ti, i, j]),
                float(v100[ti, i, j])
            ])

wind_2020_lat_long_full_df = pd.DataFrame(
    rows,
    columns=["time", "lat", "lon", "u100", "v100"]
)

display(wind_2020_lat_long_full_df)

# Save
wind_2020_lat_long_full_df.to_parquet(
    "../Data/raw-data/nc-climate-dataset/wind/processed/wind_all_2020.parquet"
)
```

## Merging 2020 - 2025 parquet data

```python
file_path_2020 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2020.parquet')
file_path_2021 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2021.parquet')
file_path_2022 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2022.parquet')
file_path_2023 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2023.parquet')
file_path_2024 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2024.parquet')
file_path_2025 = Path('../Data/raw-data/nc-climate-dataset/wind/processed/wind_2025.parquet')
```

```python
wind_2020 = pd.read_parquet(file_path_2020)
wind_2021 = pd.read_parquet(file_path_2021)
wind_2022 = pd.read_parquet(file_path_2022)
wind_2023 = pd.read_parquet(file_path_2023)
wind_2024 = pd.read_parquet(file_path_2024)
wind_2025 = pd.read_parquet(file_path_2025)
```

### Option A: Plot u100 and v100 as two separate lines

```python
import matplotlib.dates as mdates

plt.rcParams['font.family'] = 'Times New Roman'

plt.figure(figsize=(15, 8))

plt.plot(
    wind_2020_2025['u100'],
    label='U100 (East-West Wind)',
    linewidth=1.2,
    color='navy'
)

plt.plot(
    wind_2020_2025['v100'],
    label='V100 (North-South Wind)',
    linewidth=1.2,
    color='darkred'
)

plt.title('100m Wind Components (U100 & V100) from 2020 to 2025 (Hourly)', fontsize=22)
plt.xlabel('Date and Time', fontsize=18)
plt.ylabel('Wind Component (m/s)', fontsize=18)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))

plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

plt.legend(fontsize=18)
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.5, color='black')

plt.tight_layout()

plt.savefig(
    "wind_components_2020_2025.png",
    dpi=300,
    bbox_inches='tight',
    facecolor='white'
)

plt.show()
```

### Option B: Plot wind speed (derived from u100 and v100) as one line

```python
wind_2020_2025['wind_speed'] = (
    wind_2020_2025['u100']**2 + wind_2020_2025['v100']**2
)**0.5
```

```python
plt.figure(figsize=(15, 8))

plt.plot(
    wind_2020_2025['wind_speed'],
    label='Wind Speed (100m)',
    linewidth=1.2,
    color='navy'
)

plt.title('100m Wind Speed from 2020 to 2025 (Hourly)', fontsize=22)
plt.xlabel('Date and Time', fontsize=18)
plt.ylabel('Wind Speed (m/s)', fontsize=18)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))

plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)

plt.legend(fontsize=18)
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.5, color='black')

plt.tight_layout()

plt.savefig(
    "wind_speed_2020_2025.png",
    dpi=300,
    bbox_inches='tight',
    facecolor='white'
)

plt.show()
```
